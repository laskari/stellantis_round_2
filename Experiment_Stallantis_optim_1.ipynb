```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = {}

    # Initialize dp for capacity 0
    dp[0] = 0

    for i in range(n):
        new_dp = dp.copy()
        for current_capacity, current_value in dp.items():
            new_capacity = current_capacity + weights[i]
            if new_capacity <= capacity:
                new_value = current_value + values[i]
                if new_capacity not in new_dp or new_value > new_dp[new_capacity]:
                    new_dp[new_capacity] = new_value
        dp = new_dp

    # Find the maximum value in the dp dictionary
    max_value = max(dp.values())
    return max_value

# Example usage:
item_names = ['Soda', 'Chips', 'Candy', 'Water', 'Chocolate', 'Gum', 'Energy Drink', 'Popcorn', 'Granola Bar']
weights = [1.5, 1.2, 0.75, 1, 1.5, 0.5, 2, 1.85, 1.55]
values = [350, 100, 50, 500, 200, 5, 400, 150, 100]
capacity = 3

max_value = knapsack(weights, values, capacity)
print(f"The maximum value that can be put in a knapsack of capacity {capacity} is {max_value}")
```

```python
def unbounded_knapsack(weights, values, capacity):
    n = len(weights)
    dp = {0: (0, [])}  # Dictionary to store (value, items list) for each capacity

    for current_capacity in range(capacity + 1):
        for i in range(n):
            new_capacity = current_capacity + weights[i]
            if new_capacity <= capacity:
                new_value = dp[current_capacity][0] + values[i]
                if new_capacity not in dp or new_value > dp[new_capacity][0]:
                    dp[new_capacity] = (new_value, dp[current_capacity][1] + [i])

    # Find the maximum value in the dp dictionary and get the corresponding items
    max_value = max(dp.values(), key=lambda x: x[0])
    return max_value

# Example usage:
item_names = ['Soda', 'Chips', 'Candy', 'Water', 'Chocolate', 'Gum', 'Energy Drink', 'Popcorn', 'Granola Bar']
weights = [1.5, 1.2, 0.75, 1, 1.5, 0.5, 2, 1.85, 1.55]
values = [350, 100, 50, 500, 200, 5, 400, 150, 100]
capacity = 3

max_value, included_items = unbounded_knapsack(weights, values, capacity)
included_item_names = [item_names[i] for i in included_items]

print(f"The maximum value that can be put in a knapsack of capacity {capacity} is {max_value}")
print(f"Items included in the knapsack: {included_item_names}")
```
